```python
import glob as glob
import os
import shutil
trainsplit01 = '/home/muhammadbsheikh/workspace/ucf101/splits/ucfTrainTestlist/trainlist01.txt'
testsplit01 = '/home/muhammadbsheikh/workspace/ucf101/splits/ucfTrainTestlist/testlist01.txt'
datapath = '/home/muhammadbsheikh/workspace/ucf101/UCF-101'

des = '/home/muhammadbsheikh/workspace/ucf101/prep'
# creatinf file names list

mode = 'test'
fnames = []
if mode=='train':
    file = open(trainsplit01,'r').readlines()
    for line in file:
        fname = line.split(' ')[0]
        fnames.append(fname)
    drop = os.path.join(des,'train') 
elif mode=='test':
    file = open(testsplit01,'r').readlines()    
    for line in file:
        fname = line.split('\n')[0]
        fnames.append(fname)
    drop = os.path.join(des,'validation') 

print(fnames[0])

print(len(fnames))

labels = glob.glob(datapath+'/*')
files = glob.glob(datapath+'/*/*.avi')
print(len(files))
c,z,d = 0,0,0
for cat in labels:
    print(len(glob.glob(des+'/'+mode+'/*')))
    examples = glob.glob(os.path.join(cat,'*'))

    
    for example in examples:
        subfilename = example.split('UCF-101/')[1]
        #print(subfilename)
    # extract sub file name from the label   
        category = cat.split('/')[-1]
        print(subfilename)
        if subfilename in fnames:            # if subfilename is in fnames
            dest = os.path.join(drop,category,example.split('/')[-1])                                 
            if os.path.exists(os.path.join(drop,category)): # if dir exists                  
                shutil.copyfile(example,dest) # copy
                c = c + 1
            else: # else create n copy                 
                print('create')
                d = d+1
                os.mkdir(os.path.join(drop,category)) 
                dest = os.path.join(drop,category,example.split('/')[-1])                                 
                shutil.copyfile(example,dest) # copy
        else : 
            #print('#'+subfilename)    
            z = z+1   

print(c,d,z,c+z) 
```
