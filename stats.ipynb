```python
#%%

import glob as glob

train_path = '1/train/*'
train_files = glob.glob(train_path)
print(len(train_files))
list_of_file = []
class_names = []
for x in train_files:
    #print(len(glob.glob(x+'/*')))
    list_of_file.append(len(glob.glob(x+'/*')))
    class_names.append(x.split('/')[2])

print(list_of_file)
print(class_names)

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.barh(class_names[:20],list_of_file[:20],align='center')
ax.set_xlabel('clips')
ax.set_ylabel('class')
plt.show()
```
